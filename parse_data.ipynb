```python
import re

import numpy as np
import pandas as pd
```

```python
mat_file = "data/GSE4115_series_matrix.txt"

raw_mat = open(mat_file, "r").read()
raw_list = raw_mat.split("!")
```

## Auxiliary Parsing

```python
sample_titles = [re.sub(r"[\'\"\n]", "", s) for s in raw_list[33].split("\t")[1::]]
sample_ids = [int(s.split(" ")[-1]) for s in sample_titles]
sample_geoids = [re.sub(r"[\'\"\n]", "", s) for s in raw_list[34].split("\t")[1::]]
sample_chars = [re.sub(r"[\'\"\n]", "", s) for s in raw_list[42].split("\t")[1::]]
```

```python
sample_ages = [re.sub(r"[\'\"\n]", "", s.split(" ")[-1]) for s in raw_list[44].split("\t")[1::]]
sample_ages = [int(s) if s else '' for s in sample_ages]
```

```python
sample_genders = [re.sub(r"[\'\"\n]", "", s.split(" ")[-1]) for s in raw_list[45].split("\t")[1::]]
sample_races = [re.sub(r"[\'\"\n]", "", s.split(" ")[-1]) for s in raw_list[46].split("\t")[1::]]
```

```python
sample_cancer = []
for i in range(len(sample_titles)):
    if "NOT" in sample_titles[i]:
        sample_cancer.append(0)
    elif "suspect" in sample_titles[i]:
        sample_cancer.append(-1)
    else:
        sample_cancer.append(1)
```

```python
sample_status = [re.sub(r"[\'\"\n]", "", s.split(": ")[-1]) for s in raw_list[47].split("\t")[1::]]
for i in range(len(sample_status)):
    if sample_status[i] == "Cancer":
        sample_status[i] = 1
    elif sample_status[i] == "No Cancer":
        sample_status[i] = 0
    else:
        sample_status[i] = -1
```

```python
sample_quit_smoking =  [re.sub(r"[\'\"\n]", "", s.split(": ")[-1]) for s in raw_list[48].split("\t")[1::]]
```

```python
sample_packyears = [re.sub(r"[\'\"\n]", "", s.split(": ")[-1]) for s in raw_list[49].split("\t")[1::]]
sample_packyears = [float(s) if s else '' for s in sample_packyears]
```

```python
sample_hemopytsis = [re.sub(r"[\'\"\n]", "", s.split(": ")[-1]) for s in raw_list[51].split("\t")[1::]]
for i in range(len(sample_hemopytsis)):
    if sample_hemopytsis[i] == '0':
        sample_hemopytsis[i] = 0
    elif sample_hemopytsis[i] == '1':
        sample_hemopytsis[i] = 1
    else:
        sample_hemopytsis[i] = -1
```

```python
sample_lymphadenopathy = [re.sub(r"[\'\"\n]", "", s.split(": ")[-1]) for s in raw_list[52].split("\t")[1::]]
for i in range(len(sample_lymphadenopathy)):
    if sample_lymphadenopathy[i] == '0':
        sample_lymphadenopathy[i] = 0
    elif sample_lymphadenopathy[i] == '1':
        sample_lymphadenopathy[i] = 1
    else:
        sample_lymphadenopathy[i] = -1
```

```python
sample_mass_3cm = [re.sub(r"[\'\"\n]", "", s.split(": ")[-1]) for s in raw_list[53].split("\t")[1::]]
for i in range(len(sample_mass_3cm)):
    if sample_mass_3cm [i] == '0':
        sample_mass_3cm [i] = 0
    elif sample_mass_3cm [i] == '1':
        sample_mass_3cm [i] = 1
    else:
        sample_mass_3cm [i] = -1
```

```python
sample_biomarkers = [re.sub(r"[\'\"\n]", "", s.split(": ")[-1]) for s in raw_list[54].split("\t")[1::]]
sample_biomarkers = [float(s) if s else '' for s in sample_biomarkers]
```

```python
sample_risk = [re.sub(r"[\'\"\n]", "", s.split(": ")[-1]) for s in raw_list[55].split("\t")[1::]]
for i in range(len(sample_risk)):
    if sample_risk[i] == "Low":
        sample_risk[i] = 0
    elif sample_risk[i] == "Medium":
        sample_risk[i] = 1
    elif sample_risk[i] == "High":
        sample_risk[i] = 2
    else:
        sample_risk[i] = -1
```

```python
sample_data = np.array([
    sample_geoids, sample_ids, sample_titles, sample_chars, 
    sample_ages, sample_genders, sample_races,
    sample_cancer, sample_status, sample_quit_smoking, sample_packyears,
    sample_hemopytsis, sample_lymphadenopathy, sample_mass_3cm,
    sample_biomarkers, sample_risk
]).T

sample_columns = ["geoid", "id", "title", "characteristics", "age", "gender", "race",
                 "cancer", "status", "quit_smoking", "pack_years",
                 "hemopytsis", "lymphadenopathy", "mass_greater_3cm", 
                 "biomarker_score", "subjective_risk"]
```

```python
sample_df = pd.DataFrame(sample_data, columns = sample_columns)
sample_df.head()
```

```python
sample_df.to_csv("data/sample_auxiliary.csv", index = False)
```

## Series Matrix

```python
raw_matrix = raw_list[83].split("\n")
```

```python
matrix_geoids = [re.sub(r"[\'\"\n]", "", s) for s in raw_matrix[1].split("\t")[1::]]
matrix_data = raw_matrix[2:-2]
```

```python
n_geoid = len(matrix_geoids)
n_rna = len(matrix_data)

series_matrix = np.zeros((n_rna, n_geoid))
for r in range(n_rna):
    data_str = matrix_data[r]
    data_list = [float(s) for s in data_str.split("\t")[1::]]
    
    series_matrix[r] = data_list
```

```python
matrix_df = pd.DataFrame(series_matrix.T, index = matrix_geoids)
matrix_df.to_csv("data/series_matrix.csv", header = False, index = True)

matrix_df.head()
```
